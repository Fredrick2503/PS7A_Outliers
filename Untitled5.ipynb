```python
def clean_text(col):
    return (
        col.str.lower()
        .str.strip()
        .str.replace(r"[^\w\s]", "", regex=True)
        .str.replace(r"\s+", " ", regex=True)
    )


for df in [audio, info, pop]:
    df["track_name"] = clean_text(df["track_name"])
    df["artist_name"] = clean_text(df["artist_name"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_missing_values(missing_series, title):

    # Filter columns with missing values
    missing = missing_series[missing_series > 0]

    plt.figure(figsize=(12,6))

    if missing.empty:
        # No missing values case
        plt.text(
            0.5, 0.5,
            "No Missing Values Remaining",
            horizontalalignment='center',
            verticalalignment='center',
            fontsize=16,
            transform=plt.gca().transAxes
        )
        plt.title(title)
        plt.axis("off")

    else:
        ax = sns.barplot(
            x=missing.index,
            y=missing.values,
            palette="magma"
        )

        plt.title(title)
        plt.xlabel("Columns")
        plt.ylabel("Missing Value Count")
        plt.xticks(rotation=45, ha="right")

        # Add labels
        for container in ax.containers:
            ax.bar_label(container, fmt='%d', padding=3)

    plt.tight_layout()
    plt.show()
plot_missing_values(missing_before, "Missing Values Per Column (Before Cleaning)")
plot_missing_values(missing_after, "Missing Values Per Column (After Cleaning)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# -----------------------------
# 1. Select features
# -----------------------------
features = [
    "danceability",
    "energy",
    "tempo",
    "loudness",
    "valence",
    "acousticness"
]

# -----------------------------
# 2. Select top genres only
# -----------------------------
top_genres = df["genre"].value_counts().head(12).index

df_top = df[df["genre"].isin(top_genres)]

# -----------------------------
# 3. Compute mean feature values per genre
# -----------------------------
genre_profile = df_top.groupby("genre")[features].mean()

# -----------------------------
# 4. Normalize features for radar chart
# -----------------------------
scaler = MinMaxScaler()

genre_scaled = pd.DataFrame(
    scaler.fit_transform(genre_profile),
    columns=genre_profile.columns,
    index=genre_profile.index
)

# -----------------------------
# 5. Radar chart setup
# -----------------------------
genres = genre_scaled.index.tolist()
categories = genre_scaled.columns.tolist()

N = len(categories)

angles = np.linspace(0, 2*np.pi, N, endpoint=False).tolist()
angles += angles[:1]

# Determine subplot grid
cols = 4
rows = int(np.ceil(len(genres) / cols))

fig, axes = plt.subplots(
    rows,
    cols,
    subplot_kw=dict(polar=True),
    figsize=(16, rows * 4)
)

axes = axes.flatten()

# -----------------------------
# 6. Plot radar charts
# -----------------------------
for i, genre in enumerate(genres):

    values = genre_scaled.loc[genre].tolist()
    values += values[:1]

    ax = axes[i]

    ax.plot(angles, values, linewidth=2)
    ax.fill(angles, values, alpha=0.25)

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=8)

    ax.set_title(genre, fontsize=11, pad=15)

    ax.set_ylim(0,1)

# Remove unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# -----------------------------
# 7. Final formatting
# -----------------------------
plt.suptitle(
    "Audio Feature Profiles Across Music Genres",
    fontsize=16,
    y=1.02
)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from pandas.plotting import parallel_coordinates

sns.set(style="whitegrid")

# ------------------------------------------------
# Define audio features
# ------------------------------------------------
features = [
    "danceability",
    "energy",
    "tempo",
    "loudness",
    "valence",
    "acousticness"
]

# ------------------------------------------------
# 1. Genre Distribution
# ------------------------------------------------
plt.figure(figsize=(10,6))

top_genres = df["genre"].value_counts().head(10)

sns.barplot(
    x=top_genres.values,
    y=top_genres.index,
    palette="viridis"
)

plt.title("Top Genres in Dataset")
plt.xlabel("Number of Songs")
plt.ylabel("Genre")

plt.show()

print("Insight:")
print("The dataset is dominated by a small number of genres.")
print("This indicates certain musical styles are more represented in the dataset.\n")


# ------------------------------------------------
# 2. Sonic Fingerprints Across Genres
# Parallel Coordinates Plot
# ------------------------------------------------

top_genres = df["genre"].value_counts().head(8).index
genre_df = df[df["genre"].isin(top_genres)]

genre_profile = genre_df.groupby("genre")[features].mean().reset_index()

scaler = MinMaxScaler()
genre_profile[features] = scaler.fit_transform(genre_profile[features])

plt.figure(figsize=(12,6))

parallel_coordinates(
    genre_profile,
    "genre",
    colormap="viridis"
)

plt.title("Sonic Fingerprints Across Genres")
plt.ylabel("Normalized Feature Values")

plt.show()

print("Insight:")
print("Genres display distinctive sonic fingerprints across energy, tempo, and acousticness.")
print("However, several genres overlap, showing that musical styles are not always clearly separated.\n")


# ------------------------------------------------
# 3. Feature Influence on Popularity
# ------------------------------------------------

corr = df[features + ["popularity"]].corr()["popularity"].drop("popularity")

plt.figure(figsize=(8,5))

corr.sort_values().plot(
    kind="barh",
    color="steelblue"
)

plt.title("Audio Features Related to Popularity")
plt.xlabel("Correlation with Popularity")

plt.show()

print("Insight:")
print("Danceability and energy often show the strongest relationship with popularity.")
print("Acousticness tends to have weaker association with mainstream success.\n")


# ------------------------------------------------
# 4. Popularity Gradient Scatter
# ------------------------------------------------

plt.figure(figsize=(8,6))

plt.scatter(
    df["danceability"],
    df["popularity"],
    alpha=0.3,
    c=df["energy"],
    cmap="viridis"
)

plt.colorbar(label="Energy")

plt.xlabel("Danceability")
plt.ylabel("Popularity")

plt.title("Danceability vs Popularity (Energy Gradient)")

plt.show()

print("Insight:")
print("Songs with higher danceability tend to cluster around higher popularity levels.")
print("Energy further amplifies this effect, indicating upbeat songs perform well.\n")


# ------------------------------------------------
# 5. Do Hits Have a Sonic Blueprint?
# ------------------------------------------------

df["pop_group"] = pd.cut(
    df["popularity"],
    bins=[0,30,70,100],
    labels=["Deep Cuts","Moderate","Hits"]
)

profile = df.groupby("pop_group")[features].mean()

plt.figure(figsize=(10,6))

profile.T.plot(
    kind="line",
    marker="o"
)

plt.title("Average Audio Features Across Popularity Levels")

plt.ylabel("Feature Value")
plt.xlabel("Audio Features")

plt.show()

print("Insight:")
print("Hit songs tend to show slightly higher danceability and energy.")
print("However, differences are moderate, suggesting there is no single formula for popularity.\n")


# ------------------------------------------------
# 6. Genre Clustering in Feature Space
# PCA Visualization
# ------------------------------------------------

X = df[features]

pca = PCA(n_components=2)

components = pca.fit_transform(X)

pca_df = pd.DataFrame({
    "PC1": components[:,0],
    "PC2": components[:,1],
    "genre": df["genre"]
})

top_genres = df["genre"].value_counts().head(8).index
pca_df = pca_df[pca_df["genre"].isin(top_genres)]

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=pca_df,
    x="PC1",
    y="PC2",
    hue="genre",
    alpha=0.5
)

plt.title("Genre Clustering in Audio Feature Space")

plt.show()

print("Insight:")
print("Some genres cluster distinctly in audio feature space.")
print("However, significant overlap indicates that many genres share similar sonic traits.\n")
```

##Summary

The exploratory analysis reveals that the dataset is dominated by a small number of genres, with electronic, alternative, and anime containing the highest number of songs. Examination of audio feature distributions shows that most tracks cluster around moderate values of danceability and energy, with tempo centered around typical pop music ranges and loudness concentrated within a narrow band. When comparing sonic profiles across genres, distinct patterns emerge—electronic and dance genres tend to exhibit higher energy and tempo, while genres such as folk and blues show relatively higher acousticness and lower intensity. Despite these differences, there is noticeable overlap between genres, indicating that many musical styles share similar audio characteristics. Overall, the findings suggest that while genres exhibit recognizable sonic tendencies, their boundaries are not rigid, and a wide variety of audio feature combinations can coexist across musical categories.
